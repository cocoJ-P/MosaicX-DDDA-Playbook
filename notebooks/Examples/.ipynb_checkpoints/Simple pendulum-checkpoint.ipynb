### 2.1.1æ®‹å·®å½¢å¼çš„å±€é™

åœ¨æ®‹å·®å½¢å¼ä¸‹ï¼Œæ¯”å¦‚å•æ‘†

* åœ¨å•æ‘†çš„ä¾‹å­ä¸­ï¼Œ**æ®‹å·®å½¢å¼é€‚åˆé€»è¾‘åˆ¤å®šï¼ˆèƒ½å¦æ˜¾å‡½æ•°åŒ–ï¼‰ï¼Œä½†ä¸é€‚åˆæ•°å€¼è¯Šæ–­ï¼ˆdet/cond/SVDï¼‰**ã€‚

### 2.1.2 ä¸€é˜¶ç³»ç»ŸåŒ–çš„ä¼˜åŠ¿

æŠŠå®ƒè½¬ä¸ºä¸€é˜¶ç³»ç»Ÿï¼š

$$
\begin{cases}
x_1' = x_2 \\
x_2' = -\sin(x_1)
\end{cases}
$$

å¯¹åº”å‘é‡åœº

$$
f(x_1,x_2) = \begin{bmatrix} x_2 \\ -\sin(x_1)\end{bmatrix},
$$

Jacobian æ˜¯

$$
J = \begin{bmatrix}
0 & 1 \\
-\cos(x_1) & 0
\end{bmatrix}.
$$

* è¿™æ˜¯ä¸€ä¸ª **æ–¹é˜µ (2Ã—2)**ï¼š

  * $\det(J) = \cos(x_1)$ å¯ç”¨äºå¥‡å¼‚æ€§åˆ¤å®šï¼›
  * $\kappa(J)$ æ¡ä»¶æ•°åæ˜ äº†å±€éƒ¨çº¿æ€§åŒ–çš„æ•æ„Ÿæ€§ï¼›
  * SVD çš„å¥‡å¼‚å€¼ç»™å‡ºä¸»æ–¹å‘ä¸Šçš„æ”¾ç¼©å› å­ï¼Œèƒ½é‡åŒ–ç³»ç»Ÿçš„å±€éƒ¨åˆšæ€§/è„†å¼±æ€§ã€‚

æ‰€ä»¥åœ¨è¿™ä¸ªä¾‹å­ä¸­ **ä¸€é˜¶ç³»ç»Ÿè§†è§’æ›´é€‚åˆåšæ•°å€¼è¯Šæ–­ (det/cond/SVD)**ã€‚

---


```python
import sympy as sp

# --- 1) æ— é‡çº²å•æ‘†ï¼štheta'' + sin(theta) = 0 ---
tau = sp.symbols('tau')           # æ— é‡çº²æ—¶é—´
theta = sp.Function('theta')(tau)

# --- 2) ä¸€é˜¶ç³»ç»ŸåŒ– ---
x1, x2 = sp.symbols('x1 x2')      # x1 = theta, x2 = theta'
f1 = x2
f2 = -sp.sin(x1)

f = sp.Matrix([f1, f2])
x = sp.Matrix([x1, x2])

# --- 3) Jacobian ---
J = f.jacobian(x)                  # âˆ‚f/âˆ‚x
sp.pprint(J)

# å¯é€‰ï¼šç»™å‡ºæ•°å€¼ç‚¹å¤„çš„è¯„ä¼°ï¼ˆä¾‹å¦‚ x1 = pi/3, x2 = 0.2ï¼‰
J_num = J.subs({x1: sp.pi/3, x2: 0.2})
print("\nJ(pi/3, 0.2) =")
sp.pprint(J_num)
```
