```python
import math
from dataclasses import dataclass
from typing import Dict, Tuple

@dataclass
class LubricationContactParams:
    # æµä½“ä¸å‡ ä½•
    mu: float                 # åŠ¨åŠ›é»åº¦ [PaÂ·s]
    R_eff: float              # ç­‰æ•ˆåŠå¾„ R_eff = R1*R2/(R1+R2) [m]
    # æ¶¦æ»‘åŠ›å¸¸æ•°ç³»æ•°ï¼ˆé»˜è®¤å– 6Ï€ï¼›æ³•å‘æ¥è‡ª Brenner å…¸å‹å°ºåº¦ï¼Œåˆ‡å‘ç»™å‡ºå¯¹æ•°é‡çº§çš„å¯è°ƒç³»æ•°ï¼‰
    Cn: float = 6.0 * math.pi # æ³•å‘æ¶¦æ»‘ç³»æ•°ï¼Œ~ 6Ï€ Î¼ R_eff^2 * v_n / h_eff
    Ct: float = 6.0 * math.pi # åˆ‡å‘æ¶¦æ»‘ç³»æ•°ï¼Œ~ Ct Î¼ R_eff * v_t * ln(R_eff/h_eff)
    # æ¥è§¦ï¼ˆHertz + dashpotï¼‰
    kn: float = 1e5           # Hertz æ³•å‘æ¥è§¦åˆšåº¦ç³»æ•° [N/m^{3/2}]
    eta_n: float = 0.0        # æ³•å‘é˜»å°¼ç³»æ•° [NÂ·s/m]
    eta_t: float = 0.0        # åˆ‡å‘é˜»å°¼ç³»æ•° [NÂ·s/m]
    # æ‹¼æ¥ä¸æ­£åˆ™åŒ–
    delta_c: float = None     # æ¥è§¦é˜ˆå€¼ï¼ˆhâ‰ˆdelta_c å¼€å§‹æ¥è§¦ï¼‰ï¼›è‹¥ä¸º None è‡ªåŠ¨è®¾ä¸º 1e-6 * R_eff
    alpha: float = None       # æ‹¼æ¥å¸¦å®½ï¼ˆsigmoid çš„å°ºåº¦ï¼‰ï¼›è‹¥ä¸º None è‡ªåŠ¨è®¾ä¸º 0.2 * delta_c
    epsilon: float = 0.0      # æ¶¦æ»‘å¥‡å¼‚æ­£åˆ™åŒ–ï¼šh_eff = sqrt(h^2 + epsilon^2)

    def __post_init__(self):
        if self.delta_c is None:
            self.delta_c = 1e-6 * self.R_eff
        if self.alpha is None:
            self.alpha = 0.2 * self.delta_c

def lubrication_contact_force(
    h: float, vn: float, vt: float, p: LubricationContactParams,
    return_jacobian: bool = True
) -> Tuple[Tuple[float, float], Dict[str, float]]:
    """
    è®¡ç®—æ¶¦æ»‘åŠ›â€“æ¥è§¦åŠ›æ‹¼æ¥æ¨¡å‹çš„ (Fn, Ft) ä»¥åŠå¯¹ (h, vn, vt) çš„ä¸€é˜¶å¯¼æ•°ï¼ˆå¯ç”¨äº Jacobianï¼‰ã€‚
    å…¬å¼ç»“æ„ï¼š
      - æ­£åˆ™åŒ–é—´éš™ï¼šh_eff = sqrt(h^2 + epsilon^2)
      - æ¶¦æ»‘ï¼ˆæ³•å‘ï¼‰ï¼šFn_lub = Cn * mu * R_eff^2 * vn / h_eff     ï¼ˆBrenner 1961; Kim & Karrila 1991ï¼‰
      - æ¶¦æ»‘ï¼ˆåˆ‡å‘ï¼‰ï¼šFt_lub = Ct * mu * R_eff * vt * ln(R_eff/h_eff)
      - æ¥è§¦ï¼ˆæ³•å‘ï¼‰ï¼šFn_cont = kn * delta^(3/2) + eta_n * vn,  delta = max(0, delta_c - h)   ï¼ˆHertz + dashpotï¼‰
      - æ¥è§¦ï¼ˆåˆ‡å‘ï¼‰ï¼šFt_cont = eta_t * vt    ï¼ˆæ— é»ç€ç®€åŒ–ï¼Œä¾¿äºå¯å¾®ä¸æ¼”ç¤ºï¼‰
      - æ‹¼æ¥ï¼šF = w(h) * F_lub + (1 - w(h)) * F_cont
               w(h) = 1 / (1 + exp(-(h - delta_c)/alpha))

    å‚æ•°ä¸å‚è€ƒï¼š
      - Brenner, H. (1961) Chem. Eng. Sci. 16: 242â€“251.
      - Kim, S. & Karrila, S.J. (1991) Microhydrodynamics.
      - Cundall & Strack (1979) GÃ©otechnique 29: 47â€“65.ï¼ˆæ¥è§¦æ€æƒ³ï¼‰
      - Kempe & FrÃ¶hlich (2012) JFM 709: 445â€“489.ï¼ˆæ‹¼æ¥æ€æƒ³ï¼‰

    è¿”å›ï¼š
      (Fn, Ft), jac  å…¶ä¸­ jac åŒ…å«å¯¹ h, vn, vt çš„åå¯¼ï¼Œç”¨äºæ•æ„Ÿæ€§/å¯é€†æ€§åˆ†æã€‚
    """
    mu, R = p.mu, p.R_eff
    Cn, Ct = p.Cn, p.Ct
    kn, eta_n, eta_t = p.kn, p.eta_n, p.eta_t
    delta_c, alpha, eps = p.delta_c, p.alpha, p.epsilon

    # --- æ­£åˆ™åŒ–é—´éš™ ---
    h_eff = math.sqrt(h*h + eps*eps)  # C^1 å¹³æ»‘ï¼Œé¿å… 1/h å¥‡å¼‚

    # --- æ¶¦æ»‘åŠ›ï¼ˆæ˜¾å‡½æ•°ï¼‰---
    Fn_lub = Cn * mu * (R**2) * vn / h_eff
    # åˆ‡å‘æ¶¦æ»‘é‡‡ç”¨å¯¹æ•°é¡¹ï¼ˆå‡ ä½•/è¾¹ç•Œæ¡ä»¶ä¸åŒå¸¸æ•°é¡¹ä¼šå˜ï¼ŒCt ç”¨ä½œæ•´ä½“ç³»æ•°ï¼‰
    # æ³¨æ„ï¼šå½“ R<=h_eff æ—¶ ln(R/h_eff)<=0ï¼Œå®é™…ä½¿ç”¨ä¸­åº”ä¿è¯ R>>h_eff çš„è¿‘æ¥è§¦é€‚ç”¨æ€§
    # å¦‚éœ€æ›´ç¨³å¥ï¼Œå¯åŠ  max æˆ–æˆªæ–­ã€‚
    ln_term = math.log(R / h_eff) if R > h_eff else 0.0
    Ft_lub = Ct * mu * R * vt * ln_term

    # --- æ¥è§¦åŠ›ï¼ˆHertz + é˜»å°¼ï¼Œæ³•å‘ï¼›åˆ‡å‘ç®€åŒ–é˜»å°¼ï¼‰---
    delta = max(0.0, delta_c - h)
    in_contact = 1.0 if delta > 0.0 else 0.0
    Fn_cont = in_contact * (kn * (delta ** 1.5) + eta_n * vn)
    Ft_cont = in_contact * (eta_t * vt)

    # --- å¹³æ»‘æ‹¼æ¥æƒé‡ ---
    z = (h - delta_c) / alpha
    # é¿å…æº¢å‡º
    if z >= 50.0:
        w = 1.0
        dw_dh = 0.0
    elif z <= -50.0:
        w = 0.0
        dw_dh = 0.0
    else:
        w = 1.0 / (1.0 + math.exp(-z))
        dw_dh = (w * (1.0 - w)) / alpha  # sigmoid å¯¼æ•°

    # --- æ€»åŠ› ---
    Fn = w * Fn_lub + (1.0 - w) * Fn_cont
    Ft = w * Ft_lub + (1.0 - w) * Ft_cont

    if not return_jacobian:
        return (Fn, Ft), {}

    # ====== ä¸€é˜¶å¯¼æ•°ï¼ˆä¾¿äº Jacobian/æ•æ„Ÿæ€§ï¼‰ ======
    # æ¶¦æ»‘éƒ¨åˆ†å¯¼æ•°
    # d(1/h_eff)/dh = -(h)/(h_eff^3)
    dFn_lub_dh  = Cn * mu * (R**2) * vn * ( - h / (h_eff**3) )
    dFn_lub_dvn = Cn * mu * (R**2) / h_eff

    # ln(R/h_eff) å¯¹ h çš„å¯¼æ•°ï¼š d ln(R/h_eff)/dh = - h / h_eff^2
    dFt_lub_dh  = Ct * mu * R * vt * ( - h / (h_eff**2) )
    dFt_lub_dvt = Ct * mu * R * ln_term

    # æ¥è§¦éƒ¨åˆ†å¯¼æ•°ï¼ˆå¿½ç•¥ Heaviside åœ¨ h=delta_c çš„åˆ†å¸ƒå¯¼æ•°ï¼Œåªåœ¨åŒºåŸŸå†…ç»™å‡ºï¼‰
    if in_contact > 0.0:
        # d delta / dh = -1
        dFn_cont_dh  = -1.5 * kn * math.sqrt(delta)   # = kn * (3/2)*delta^(1/2) * (-1)
        dFn_cont_dvn = eta_n
        dFt_cont_dh  = 0.0
        dFt_cont_dvt = eta_t
    else:
        dFn_cont_dh  = 0.0
        dFn_cont_dvn = 0.0
        dFt_cont_dh  = 0.0
        dFt_cont_dvt = 0.0

    # æ‹¼æ¥é“¾å¼æ³•åˆ™ï¼šF = w*F_lub + (1-w)*F_cont
    # dF/dh = dw/dh * (F_lub - F_cont) + w*dF_lub/dh + (1-w)*dF_cont/dh
    dFn_dh  = dw_dh * (Fn_lub - Fn_cont) + w * dFn_lub_dh + (1.0 - w) * dFn_cont_dh
    dFn_dvn =            w * dFn_lub_dvn                 + (1.0 - w) * dFn_cont_dvn

    dFt_dh  = dw_dh * (Ft_lub - Ft_cont) + w * dFt_lub_dh + (1.0 - w) * dFt_cont_dh
    dFt_dvt =            w * dFt_lub_dvt                 + (1.0 - w) * dFt_cont_dvt

    jac = {
        "dFn_dh": dFn_dh, "dFn_dvn": dFn_dvn,
        "dFt_dh": dFt_dh, "dFt_dvt": dFt_dvt,
        # å¦‚éœ€äº¤å‰åå¯¼ï¼ˆdFn/dvt, dFt/dvnï¼‰è¿™é‡Œä¸º 0ï¼Œå› ä¸ºæ¨¡å‹ä¸­æœªç›´æ¥ç›¸å…³
        "dFn_dvt": 0.0, "dFt_dvn": 0.0,
        # ä¹Ÿå¯å›ä¼ ä¸­é—´é‡ï¼Œä¾¿äºåˆ†æ
        "w": w, "dw_dh": dw_dh, "h_eff": h_eff, "in_contact": in_contact,
        "Fn_lub": Fn_lub, "Fn_cont": Fn_cont, "Ft_lub": Ft_lub, "Ft_cont": Ft_cont
    }
    return (Fn, Ft), jac

# ç¤ºä¾‹å‚æ•°
p = LubricationContactParams(
    mu=1.0e-3,       # æ°´çš„é»åº¦ ~1e-3 PaÂ·s
    R_eff=5.0e-4,    # 0.5 mm
    kn=1.0e6,        # ä¾‹å­ä¸­çš„æ¥è§¦åˆšåº¦
    eta_n=1.0e-3,    # æ³•å‘é˜»å°¼
    eta_t=5.0e-4,    # åˆ‡å‘é˜»å°¼
    epsilon=1.0e-8   # æ¶¦æ»‘æ­£åˆ™åŒ–å°ºåº¦ï¼ˆå¯å–ç²—ç³™åº¦/æ»‘ç§»é•¿åº¦é‡çº§ï¼‰
)
(h, vn, vt) = (2.0e-6, -0.01, 0.005)  # é—´éš™2Î¼mï¼Œæ³•å‘é è¿‘(-)ï¼Œåˆ‡å‘é€Ÿåº¦0.005 m/s

(Fn, Ft), J = lubrication_contact_force(h, vn, vt, p, return_jacobian=True)
print(f"Fn={Fn:.6e} N, Ft={Ft:.6e} N")
print("Jacobian (partials):", {k: f"{v:.3e}" for k, v in J.items() if k.startswith("d")})

```
